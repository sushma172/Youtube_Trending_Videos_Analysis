*  ######  what are the most common words in trending video titles?

```python
def most_common(title_words):
    d={}
    for i in title_words:
        d[i]=title_words.count(i)
    return d
```

```python
df1=df['title'].head(10)
title_words = list(df1.apply(lambda x: x.split()))
title_words = [x for y in title_words for x in y]
dic=most_common(title_words)
dic
```

# Data Visualization

*  ######  Which channels had the largest number of trending videos? Plot it using bar plot?

```python
cdf = df.groupby("channelTitle").size().reset_index(name="video_count").sort_values("video_count", ascending=False).head(20)
cdf.reset_index()

plt.figure(figsize=(8,8))
sns.barplot(x="video_count", y="channelTitle", data=cdf, palette=sns.cubehelix_palette(n_colors=20,reverse=True))
plt.xlabel('No of videos')
plt.ylabel('Channel')
plt.show()
```

*  ######  Which days of the week had the largest number of trending videos? Plot it using count plot?

```python
df['Days']=pd.to_datetime(df['publishedAt'],format='%Y-%m-%d %H:%M:%S')
df['Days']=df['Days'].dt.day_name()
plt.figure(figsize=(8,8))
sns.countplot(x='Days',data=df,linewidth=5,edgecolor=sns.color_palette('dark',10),palette='plasma')
plt.xlabel('Days')
plt.ylabel('No. of videos')
plt.title('Video Uploads by Day of the Week')
plt.show()
```

*  ######  Draw a pie chart to show the percentage of channelTitles based on videocount?

```python
cdf = df.groupby("channelTitle").size().reset_index(name="video_count").sort_values("video_count", ascending=False).head(5)
channel=cdf['channelTitle'].to_list()
count=cdf['video_count'].to_list()
plt.pie(count,labels=channel,colors=['red','blue','g','m','c'],autopct="% 1.2f%%",
                            explode=(0.2,0,0,0,0),shadow=True)
plt.title("Top 5 Channels by Video Count")
plt.show()
```

*  ###### How many trending video titles contain capitalized word? Visualize it using piechart.

```python
def contains_capitalized_word(s):
    for w in s.split():
        if w.isupper():
            return True
    return False


df["contains_capitalized"] = df["title"].apply(contains_capitalized_word)

value_counts = df["contains_capitalized"].value_counts().to_dict()
plt.pie([value_counts[False], value_counts[True]], labels=['No', 'Yes'], 
           colors=['#003f5c', '#ffa600'], textprops={'color': '#040204'}, startangle=45)

plt.title('Title Contains Capitalized Word?')
```

*  ###### Create the correlation table between our dataset variables. Visualize it using heatmap.

```python
df.corr()
```

```python
h_labels = [x.replace('_', ' ').title() for x in 
            list(df.select_dtypes(include=['number', 'bool']).columns.values)]

plt.subplots(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, xticklabels=h_labels, yticklabels=h_labels,linewidths=3, linecolor='purple', cmap=sns.cubehelix_palette(as_cmap=True))
plt.title('Correlation Heatmap')
plt.show()
```
